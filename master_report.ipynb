# Classification

## Full Model (Linear Regression)

```python
# Partition explanatory and response variables
X = df.drop(columns=['edjefe', 'edjefa', 'dependency', 'v2a1', 'v18q1', 'rez_esc', 'meaneduc', 'SQBmeaned'])
X = X.drop(columns=['Target', 'Id', 'idhogar'])
print(X.shape)

y = df['Target']
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=y)
```

```python
clf = LinearRegression()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
y_pred = X_pred.round()
y_pred = X_pred.astype(int)

df_test = pd.DataFrame([y_pred, y_test], index=['ypred', 'ytest'])
df_test = df_test.transpose()
df_test['correct'] = df_test['ypred'] == df_test['ytest']
df_test.correct.mean()
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test, y_pred, average='macro')
```
